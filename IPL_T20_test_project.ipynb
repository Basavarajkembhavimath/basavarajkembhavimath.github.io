```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
```

```python
def get_pl_name(x):
    pl_name = x.split('(')
    return pl_name[0]
```

```python
def get_team(x):
    team_name = re.search('\(([^)]+)', x).group(1)
    return team_name
```

```python
def indian(x):
    if str(x) == 'True':
        return 'indian'
    else:
        return 'foreign'
```

```python
def is_opener(x):
    if int(x) <= 2:
        return 'Yes'
    else:
        return 'No'
```
