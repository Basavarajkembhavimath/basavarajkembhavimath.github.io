## Reading CSV file from pandas

```python
col_names = ["col1", "col2", "col3", "col4", "col5","col6", "col7", "col8", "col9", "col10","col11"]
df1 = pd.read_csv('cricket_project.csv', names=col_names)
df1.head()
```

```python
df1.head()
```

## Creating list and adding data to it

```python
Innings = df1['col2'][11:].tolist()
Date = df1['col1'][11:].apply(get_date)
Season= df1['col1'][11:].apply(get_val)
Team_Name = df1['col4'][11:].tolist()
Balls = df1['col3'][11:].tolist()
Tournament = df1['col1'][11:].apply(get_turny)
Match_No = df1['col1'][11:].apply(get_no)
City = df1['col1'][11:].apply(get_city)
Stadium = df1['col1'][11:].apply(get_stadium)
Toss_winner = df1['col1'][11:].apply(get_ts_win)
Elected = df1['col1'][11:].apply(elected)
On_Strike = df1['col5'][11:].tolist()
Player2 = df1['col6'][11:].tolist()
Bowler = df1['col7'][11:].tolist()
Runs = df1['col8'][11:].tolist()
wide_runs = df1['col9'][11:].tolist()
Out = df1['col10'][11:].tolist()
Wkt_taken_by = df1['col11'][11:].tolist()
```

## Creating DataFrame

```python
df = pd.DataFrame({'Season': Season,
                   'Date': Date,
                   'Innings': Innings,
                   'Match_No': Match_No,
                   'Team': Team_Name,
                   'Tournament': Tournament,
                   'City': City,
                   'Stadium' : Stadium,
                   'Toss_winner': Toss_winner,
                   'Elected_To': Elected,
                   'On_Strike': On_Strike,
                   'Player2' : Player2,
                   'Bowler': Bowler,
                   'Balls': Balls,
                   'Runs' : Runs,
                   'Wide_runs': wide_runs,
                   'Wicket': Out,
                   'Wkt_taken_by': Wkt_taken_by,
                  })
```

## converting string to integer

```python
df['Innings'] = df['Innings'].astype(int)
df['Balls'] = df['Balls'].astype(float)
df['Runs'] = df['Runs'].astype(int)
df['Wide_runs'] = df['Wide_runs'].astype(int)
df['Match_No'] = df['Match_No'].astype(int)
```

### Adding one more column to data frame

```python
df['Over'] = 1 + df['Balls'].astype(int)
```
