```
import requests
from bs4 import BeautifulSoup as bs
```

```
#Load the content of webpage
r = requests.get('https://www.bbc.com/')
r1 = requests.get('https://www.bbc.com/sport/football/56150298')

#convert to a beautiful soup object
bbc = bs(r.content)
football = bs(r.content)
#print out our html
bbc.prettify()
football.prettify()
```
